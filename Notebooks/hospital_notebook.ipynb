```python
import pandas as pd
import numpy as np
```

```python
caremembers = pd.read_csv('../data/NSS_GreatCare_Members.csv')
inpatient = pd.read_csv('../data/NSS_Hospital_Inpatient_Admissions.csv')
primarycare = pd.read_csv('../data/NSS_PrimaryCare_Office_Visits.csv')
```

```python
inpatient
```

```python
dupes = inpatient.duplicated(subset= 'Patient ID', keep = False)
dupes = dupes.reset_index()
dupes = dupes.rename(columns = {0: 'readmitted'})
```

```python
dupes
```

```python
readmissions = inpatient[['Patient ID', 'Encounter Start Date', 'Encounter End Date']]
readmissions = readmissions.reset_index()
readmissions = pd.merge(readmissions, dupes, on = 'index')
```

```python
readmitted = readmissions[readmissions['readmitted'] == True]
```

```python
readmitted
```

```python
readmitted['Encounter Start Date'] = pd.to_datetime(readmitted['Encounter Start Date'])
readmitted['Encounter End Date'] = pd.to_datetime(readmitted['Encounter End Date'])
readmitted
```
